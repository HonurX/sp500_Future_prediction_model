```python
from prophet import Prophet
```

```python
from prophet.diagnostics import cross_validation
from prophet.diagnostics import  performance_metrics
from prophet.plot import plot_cross_validation_metric
```

```python
df = df.rename(columns={"Year" : "ds"  , "y" : "AVP"})
```

```python
df["ds"] = pd.to_datetime(df["ds"],format="%Y")
```

```python
df.columns.values[6] ="Annual Cng" ##way to change duplicated column name by numeric
```

```python
df.columns.values[1] = "y"
```

```python
df
```

```python
df = df.rename(columns={"y" : "AVG"})
```

```python
df
```

```python
df = df.rename(columns={"Year High" : "y"})
```

```python
df
```

```python
prop_model = Prophet()
```

```python
prop_model.fit(df[["ds" , "y"]])
```

```python
future = prop_model.make_future_dataframe(periods=24 , freq="YE")
```

```python
forecast = prop_model.predict(future)
```

```python
forecast[["ds" , "yhat" , "yhat_lower" , "yhat_upper"]].tail(24)


```

```python
plt.figure(figsize=(20,10))
prop_model.plot(forecast)
plt.title("SP500 FORECAST")
plt.show()
plt.close()
```
